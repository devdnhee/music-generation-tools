```python
import os

test_input_path = "../data/stems/ylwib_0000_0200/vocals.wav"
test_output_path = "../data/midi/ylwib_0000_0200/"

os.makedirs(os.path.dirname(test_output_path), exist_ok=True)

!basic-pitch $test_output_path $test_input_path
```

```python
import pygame


def play_midi_file(midi_file_path):
    # Initialize the pygame mixer
    pygame.init()
    pygame.mixer.init()

    pygame.mixer.music.load(midi_file_path)
    pygame.mixer.music.play()
    # Wait for the music to finish playing
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)
    # Stop the mixer
    pygame.mixer.music.stop()
    pygame.mixer.quit()


# Play the generated MIDI file
out_file_path = os.path.join(
    test_output_path,
    f"{os.path.basename(test_input_path).split('.')[0]}_basic_pitch.mid",
)
play_midi_file(out_file_path)
```
